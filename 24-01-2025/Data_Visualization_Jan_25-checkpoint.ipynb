```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,5,6,9,13]

plt.plot(x,y,linestyle='dashed',color='hotpink',linewidth=2,marker='*')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[23000,40000,11000,70000,25000,90000]
plt.plot(x,y,linestyle='dashed',color='green',marker='*')
plt.xlabel('Months')
plt.ylabel('Savings Amount')
plt.title('Growth of savings account over 6 months')
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]

plt.bar(categories,values,color=['brown','purple','green','magenta','skyblue'],edgecolor='black')

plt.title('Basic Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

```python
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,37,41,42,45,46,49,52,54,57,58,62,63,65,68,69]
age=[23,46,68]
b=[20,30,40,50,60,70]

#bins:Defines the intervals for grouping the data
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title('Age Distribution of Customers')
plt.xlabel('Age Ranges')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
languages=['Python','Ruby','C++','Java']
percentage=[40,20,10,30]

plt.pie(percentage,labels=languages,colors=['orange','hotpink','lightgreen','skyblue'],autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.legend()
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]

plt.scatter(categories,values,color=['brown','purple','green','magenta','skyblue'],s=100)

plt.title('Basic Scatter Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```
