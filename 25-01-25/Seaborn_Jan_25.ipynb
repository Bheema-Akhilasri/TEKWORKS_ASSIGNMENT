```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#HISTOGRAM PLOT
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color='skyblue')
plt.title('Distribution of Total Bills')
plt.show()
```

```python
#KDE PLOT KERNAL DENSITY DISTRIBUTION - visualizes the probability density of a variable.
sns.kdeplot(data=tips,x='total_bill',color='pink',fill=True)#fill instead of shade
plt.title("KDE Plot of Total Bills")
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',palette='viridis')
plt.title('Average Total Bill by Day')
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',palette='coolwarm')
plt.title('Box Plot of Total Bills by Day and Gender')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',palette='deep')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
#Line Plot
sns.lineplot(data=tips,x='size',y='tip',marker='o',color='green')
plt.title('Line PLot of Party Size vs Tip')
sns.pairplot(data=tips,hue='sex',palette='husl')
plt.show()
```

```python
#Heatmap
numeric_tips = tips.select_dtypes(include=[float, int]) 
corr_matrix=numeric_tips.corr()
sns.heatmap(corr_matrix,annot=True,cmap='rainbow',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
survival=pd.read_csv('train.csv')
survival.head()
```

```python
sns.barplot(data=survival,x='Pclass',y='Survived',palette='Pastel1',errorbar=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data=survival,x='Age',y='Fare',hue='Sex',palette='deep')
plt.title('Relation between Age and Fare')
plt.show()
```

```python
temp=survival[['Age','Fare','Pclass','Survived']]
sns.pairplot(data=temp,hue='Survived',palette='husl',diag_kind='kde')
plt.suptitle('Pairwise Relationships',y=1.02)
plt.show()
```

```python
numeric_tips = survival.select_dtypes(include=[float, int]) 
corr_matrix=numeric_tips.corr()
sns.heatmap(corr_matrix,annot=True,cmap='Set2',fmt='.2f',linewidth=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
