```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('parkinsons.data')
df
```

```python

df.drop(columns=['name'], inplace=True)
X = df.drop(columns=['status'])  
y = df['status']  

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['status'] = y

plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='status', palette=['blue', 'red'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Parkinson’s Disease vs Healthy')
plt.legend(title="Status", labels=["Healthy", "Parkinson's"])
plt.grid()
plt.show()
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
plt.figure(figsize=(8, 5))
plt.bar(range(1, 3), explained_variance, alpha=0.7, label='Explained Variance')
plt.plot(range(1, 3), cumulative_variance, marker='o', linestyle='--', color='r', label='Cumulative Variance')
plt.xticks([1, 2])
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('Explained Variance Ratio')
plt.legend()
plt.show()
```
