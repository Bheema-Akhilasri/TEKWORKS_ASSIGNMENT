FEATURES
1) Product_ID : Unique identifier for each product
2) Sale_Date : The date on which the sale occurred
3) Sales_Rep : Name of the sales representative handling the transaction
4) Region : Geographic location where the sale took place
5) Sales_Amount : The total revenue generated from the sale
6) Quantity_Sold : The number of units sold
7) Product_Category : Category of the product sold (e.g., Electronics, Clothing, Furniture)
8) Unit_Cost : The cost of a single unit of the product
9) Unit_Price : The price at which a single unit was sold

10) Customer_Type : Indicates whether the customer is "New" or "Returning"
11) Discount : Discount applied to the sale, represented as a decimal (e.g., 0.07 = 7%)
12) Payment_Method : The mode of payment used (e.g., Credit Card, Cash, Bank Transfer)
13) Sales_Channel : The platform through which the sale was made (e.g., Online, Retail)
14) Region_and_Sales_Rep : A combined field of the sales region and sales representative

```python
sales['Year']=sales['Sale_Date'].str.split('-').str[0]
sales['Month']=sales['Sale_Date'].str.split('-').str[1]
sales['Date']=sales['Sale_Date'].str.split('-').str[2]
```

```python
sales.info()
```

```python
sales['Year']=sales['Year'].astype(int)
sales['Month']=sales['Month'].astype(int)
sales['Date']=sales['Date'].astype(int)
```

```python
sales.info()
```

```python
sales.drop('Sale_Date',axis=1,inplace=True)
```

```python
sales.drop('Sales_Rep',inplace=True,axis=1)
```

```python
sales.drop('Region_and_Sales_Rep',inplace=True,axis=1)
```

```python
encoder = OneHotEncoder(sparse_output=False, drop='first')  
categorical_cols = ['Region','Product_Category','Customer_Type','Payment_Method','Sales_Channel']
encoded_cols = pd.DataFrame(
    encoder.fit_transform(sales[categorical_cols]),
    columns=encoder.get_feature_names_out(categorical_cols) 
)
encoded_cols.index = sales.index  
sales.drop(columns=categorical_cols, inplace=True)
sales = pd.concat([sales, encoded_cols], axis=1)
print(sales.head())
```
